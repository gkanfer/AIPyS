```python
from IPython.display import clear_output

dir1 = r'E:\Gil\Images\012224_images40x_training\WT'
files = [f for f in os.listdir(dir1) if not f.startswith('._')]
random.shuffle(files)
def runSingleCell(inPath = r'E:\Gil\Images\012224_images40x_training\WT',outPath = r'E:\Gil\Images\012224_images40x_training\input_sc_mix', filesName = files,FilesCounter = 40,CellCounter = 5000,clean = 500,lable = 'WT'):
    count = 0
    num  = 318
    while (count < FilesCounter) or (num >= CellCounter):
        clear_output(wait=True)
        AIPS_pose_object = AC.AIPS_cellpose(Image_name = files[count], path = dir1, model_type = 'cyto', channels=[0,0], clean = clean) 
        img = AIPS_pose_object.cellpose_image_load()
        mask, table = AIPS_pose_object.cellpose_segmantation(image_input=img)
        name = files[count].split(".")[0]
        count += 1
        print(f'File number : {count}')
        if (len(table) > 5) or (len(table) < 400):
            num += len(table)
            print(f'Total Number of cells: {num}')
            for i in range(len(table)):
                stack, _ = AIPS_pose_object.stackObjects_cellpose_ebimage_parametrs_method(image_input=img,
                                                                               extract_pixel=50,
                                                                               resize_pixel=150,
                                                                               img_label=table.index.values[i])
                if stack is None:
                    continue
                else:
                    plt.imsave(os.path.join(outPath,lable + '__' + f'{i}' + '__' + name +  '.png'), stack)
                    print('-' * i, end='\r')
runSingleCell(inPath = r'E:\Gil\Images\012224_images40x_training\WT', outPath = r'E:\Gil\Images\012224_images40x_training\input_sc_mix', filesName = files,FilesCounter = 40,CellCounter = 5000,clean = 500,lable = 'WT')       
```
