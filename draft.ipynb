```python
import os
from pathlib import Path
import argparse
import subprocess
import h5py
import numpy as np

def unpack_dataset(dataset):
    """Handle decoding of datasets from HDF5, with special handling for arrays of strings."""
    if isinstance(dataset, np.ndarray) and dataset.dtype.kind == 'O':  # Object arrays, potentially strings
        return dataset[0].decode('utf-8')  # Decode each string from bytes
    elif isinstance(dataset, np.ndarray) and dataset.dtype.type is np.bytes_:  # Direct array of bytes
        return np.char.decode(dataset, 'utf-8')  # Decoding bytes array to strings
    elif isinstance(dataset, bytes):  # Single item as bytes
        return dataset.decode('utf-8')  # Decode single byte string to normal string
    else:
        return dataset  # Return as is for other datatypes




# Assuming 'required_params' and 'option' are defined as per your description
required_params = {
    "measDia": ["Image_name"],
    "cp_seg_video": ["Image_name", "videoName", "model_type", "channels", "outPath"],
    "cp_gran_video": ["start_kernel", "end_kernel", "kernel_size", "extract_pixel", "resize_pixel", "videoName", "model_type", "channels", "Image_name", "outPath"],
    "cp_gran_table_gen": ["kernelGran", "w", "h", "extract_pixel", "resize_pixel", "videoName", "model_type", "channels", "Image_name", "outPath"],
    "dataLabeling": ["dataPath", "imagePath"],
    "data_viz": ["dataPath", "imagePath"],
    "modelBuild": ["dataPath", "imW", "imH", "thold", "extract_pixel", "resize_pixel", "videoName", "model_type", "channels", "Image_name", "outPath"],
    "deployBuild": ["kernelGran", "dataPath", "imW", "imH", "thold", "extract_pixel", "resize_pixel", "videoName", "model_type", "channels", "Image_name", "outPath"]
}

# User-selected option
option = "measDia"

# Opening the HDF5 file
with h5py.File('default_parameters.h5', 'r') as h5file:
    sub_data_dict = {}
    for subkey in required_params[option]:
        try:
            value = unpack_dataset(h5file[subkey][()])
            sub_data_dict[subkey] = value
        except KeyError:
            raise f" {subkey} is required. use set_parameters --{subkey} to update parameters"
        # test if the parameter is missing
        if value=='None':
            raise f" {subkey} is required. use set_parameters --{subkey} to update parameters"

        
    
#     # Check if the option exists in the file
#     if required_params[option][0] in h5file:
#         # Iterate over the required params for the selected option
#         for subkey in required_params[option]:
#             # Try to retrieve and store the value from the HDF5 file
#             try:
#                 value = h5file[subkey][()]
#                 sub_data_dict[subkey] = value
#             except KeyError:
#                 print(f"Warning: '{subkey}' not found in '{option}'")
#     else:
#         print(f" {subkey} is required. use set_parameters --{subkey} to update parameters")

# # Printing the resulting sub-dictionary
# print(sub_data_dict)
```

```python
import os
import shutil
import h5py
from pathlib import Path
import numpy as np
import argparse
import pdb

user_parameters_path = os.path.join(Path.home(), '.AIPyS', 'parameters.h5')
def update_user_parameters(parameter_updates):
    """Example function to update parameters in the user's parameters.h5 file."""
    with h5py.File(user_parameters_path, 'r+') as hdf:
        for key, value in parameter_updates.items():
            if key in hdf:
                value = 'None' if value is None else value
                if isinstance(value, str):
                    hdf[key][()] = np.array(value, dtype=h5py.special_dtype(vlen=str))
                else:
                    hdf[key][...] = value
            else:
                # If the key does not exist in the HDF5 file, create a new dataset.
                if isinstance(value, str):
                    dt = h5py.special_dtype(vlen=str)
                    hdf.create_dataset(key, data=np.array(value, dtype=dt), dtype=dt)
                else:
                    hdf.create_dataset(key, data=value)
                print(f"'{key}' added to or updated in the HDF5 file with value: {value}")

parameter_updates = {"areaSel":1000}
update_user_parameters(parameter_updates)
```
